```python
#4.Univariate Analysis: for both categorical columns and numerical columns
def univar_analysis(df):
    col=df.columns
    for i in col:
        if df[i].dtype=='int32' or df[i].dtype=='int64' or df[i].dtype=='float':
            print("__________________________________UNIVARIATE ANALYSIS of NUMERICAL COLUMN {}_________________________________".format(i))
            maximum=df[i].max()
            minimum=df[i].min()
            avg=df[i].mean()
            var=df[i].var()
            std=df[i].std()
            total=df[i].sum()
            print("**********************************Data Analysis of {} using Non-Visualization************************************".format(i))
            print("Maximum value in {} is : {}".format(i,maximum))
            print("Minimum value in {} is : {}".format(i,minimum))
            print("Average/mean value of {} is : {} ".format(i,avg))
            print("Variance of {} is : {}".format(i,var))
            print("Standard Deviation of {} is : {}".format(i,std))
            print("Total of the values in {} is : {}".format(i,total))
            print("**********************************Data Analysis of {} using Visualization*****************************************".format(i))
            plt.hist(x=df[i])
            plt.title('Histogram showing the distribution of IMDB top-rated movies by {}'.format(i))
            plt.xlabel(i)
            plt.ylabel('No.of.movies')
            plt.show()
            plt.boxplot(x=df[i])   #all the numerical figures
            plt.title('Box plot showing outliers in {}'.format(i))
            plt.show()
            
        else:
            unique=df1[i].unique()
            nunique=df[i].nunique()
            value_counts=df[i].value_counts()
            print('_____________________________UNIVARIATE ANALYSIS of CATEGORICAL COLUMN {}______________________________'.format(i))
            print('Number of Distinct values in column {} are {}'.format(i,nunique))
            if nunique<30:    #to avoid categorical columns with high cardinality
                print("***********************************Data Analysis of {} using non-visualization******************************".format(i))
                print('Distinct values in {} are:\n {} '.format(i,unique))
                print('Frequency of each data/value in the column {}:\n {}'.format(i,value_counts))
                  #go in this specified order only for plotting with plt
                plt.figure(figsize=(20,10))   
                plt.bar(df[i].value_counts().index,df[i].value_counts().values) 
                plt.xticks(rotation=90)
                plt.title("Bar chart showing frequency of each {}".format(i),fontsize=20)
                print("***********************************Data Analysis of {} using visualizaion***********************************".format(i))
                plt.show()
                plt.figure(figsize=(60,30))
                plt.pie(x=df[i].value_counts(),labels=df[i].value_counts().index,autopct='%0.2f%%')
                plt.title("Pie chart showing the proportion of each {} as a total".format(i),fontsize=20)
                plt.show()
            else:
                print("Because of high cardinality, cannot perform meaningful analysis for column {}".format(i))
            
            
            
           
```

```python
univar_analysis(df1) #applying custom univariate anaylsis function on our dataset df1
```
